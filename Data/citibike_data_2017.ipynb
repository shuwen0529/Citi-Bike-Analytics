```python
import glob
import pandas as pd
import numpy as np
```

```python
clean_data_ridership = pd.concat([pd.read_csv(f) for f in glob.glob('raw_data/CitiBikeDaily/*.csv')],\
                                 ignore_index = True)
clean_data_ridership.head()
```

```python
clean_data_ridership.count()
```

```python
li = []
for i in range(1, 13):
    path = f"raw_data/CitiBikeTripHistories/JC-2017{i:02d}-citibike-tripdata.csv" 
    df = pd.read_csv(path, index_col=None, header=0)
    li.append(df)
clean_data_trip = pd.concat(li, axis=0, ignore_index=True)
clean_data_trip.tail()
```

```python
clean_data_trip.count()
```

```python
# Export to csv
clean_data_ridership.to_csv("cleaned_data/clean_quarter_data_2017.csv")
clean_data_trip.to_csv("cleaned_data/clean_data_2017.csv") 
```
